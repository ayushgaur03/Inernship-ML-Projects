```
movies=pd.read_csv("movies (1).csv",error_bad_lines=False, encoding='latin-1')
movies
```

```
ratings=pd.read_csv("ratings (1).csv",error_bad_lines=False, encoding='latin-1')
ratings
```

```
links=pd.read_csv("links (1).csv",error_bad_lines=False, encoding='latin-1')
links
```

```
ratings.drop('timestamp',axis=1,inplace=True)
ratings
```

```
a = ratings['userId'].value_counts()
```

```
b = ratings['userId'].value_counts()>100
```

```
new_rating=list(a[b].index)
```

```
len(new_rating)
```

```
ratings=ratings[ratings['userId'].isin(new_rating)]
ratings.shape
```

```
movie_ratings= ratings.merge(movies,on='movieId')
movie_ratings.head()
```

```
movie_rate_link= pd.merge(links,movie_ratings,on='movieId')
movie_rate_link.head()
```

```
movie_rate_link
```

```
movie_rate_link= movie_rate_link.drop_duplicates(subset=['userId','title'])
movie_rate_link.shape
```

```
imdb = movie_rate_link.groupby('imdbId')['rating'].count().reset_index()
imdb
```

```
final = movie_rate_link[movie_rate_link['rating'].isin(imdb.index)]
final.shape
```

```
pivot=final.pivot_table(columns='imdbId',index='title',values='rating')
pivot.fillna(0,inplace=True)
pivot
```

```
from scipy.sparse import csr_matrix
movie_matrix=csr_matrix(pivot)
```

```
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```
model.fit(movie_matrix)
```

```
pivot.index[200]
```

```
pivot.iloc[200,:].values.reshape(1,-1)
```

```
distance,suggestion=model.kneighbors(pivot.iloc[1,:].values.reshape(1,-1),n_neighbors=5)
suggestion
```

```
pivot.index[50]
```

```
def recommend(moviename):
    m_id=np.where(pivot.index==moviename)[0][0]
    distance,suggestion=model.kneighbors(pivot.iloc[m_id+1,:].values.reshape(1,-1),n_neighbors=5)
    for i in suggestion[0]:
        print(pivot.index[i])
```

```
recommend('13 Going on 30 (2004)')
```

```
recommend('Jumanji (1995)')
```

```
recommend('Four Rooms (1995)')
```
